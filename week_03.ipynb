### 논리연산자
- 비교연산자가 여러번 사용될 때 사용
- and, or, not이 있음
- a < x < b(a, b는 숫자) 형태의 논리연산자는 파이썬에서만 사용 가능

```python
# and
# A and B 또는 A & B의 형태로 사용
# A와 B가 모두 참인 경우 True

a = True
b = True
if a and b: #if a: 는 a가 True인지 묻는 형태, a == True는 사용x
    print(1)
```

```python
# or
# A or B 또는 A | B의 형태로 사용
# A나 B 중 하나 이상이 참이면 True

c = False

# 둘 중 하나만 참
if a or c:
    print(1)

# 둘다 참
if a or b:
    print(2)
```

```python
# not
# not A의 형태로 사용
# A가 False면 True

if not c:
    print(1)

# a가 True기 때문에 False
if not a:
    print(2)
```

```python
# 가독성을 위해 괄호를 넣을 수 있음
a = 10
b = 5

if (a > 0) and (b > 0):
    print(1)

# 파이썬에서만 가능한 형태 A < x < B

if 0 < b < 10:
    print(2)
```

### Escape charachter(이스케이프 문자)
- 특수 문자열
- \n, \t, \b
- 이스케이프 문자가 아닌 \n 자체를 출력하고 싶다면
    - \\n : \를 하나 더 붙여서 이스케이프 문자를 무효화
    - r'' (raw string)
- 문자열의 표면상의 형태를 바꿈

```python
# \n
# 줄바꿈 효과
sentence = 'I eat \nan apple.'
print(sentence)
```

```python
# \t
# 탭이 들어감
sentence = 'I eat \tan apple.'
print(sentence)
```

```python
# \b
# 백슬래시 효과, 잘 사용하지 않음
sentence = 'I eat \ban apple.'
print(sentence)
```

```python
# \를 하나 더 쓰거나 r''로 이스케이프 문자를 무효화
sentence1 = 'I eat \\nan apple.'
print(sentence1)
sentence2 = r'I eat \nan apple.'
print(sentence2)
```
