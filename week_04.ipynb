#### list comprehension
- 포함형/지능형/축약형 리스트
- 리스트 안에서 for문이나 if문 등을 이용하여 리스트를 생성하는 것
- list + for loop
- [A for B in C]의 형태
  - C에 있는 B를 통해 A를 리스트에 담아서 반환
  - C: 반복 가능한 자료형(리스트, 문자열 등)
  - B: C에서 가져온 요소(리스트의 원소, 문자열의 글자 하나 등)
  - A: B를 이용해 도출된 값
- [A for B in C if ...]처럼 if문을 사용해 조건을 추가할 수 있음
  - 만약 ...이라면 C에 있는 B를 통해 A를 리스트에 담아서 반환

```python
[element[0] for element in animals]
```

```python
#text1의 토큰 처음 10개 중 마지막 알파벳만 반환
[word[-1] for word in text1.tokens[:10]]
```

### NLTK - FreqDist()
- Frequency Distribution: 단어의 빈도수(도수 분포)
- 단어의 개수나 빈도수 등 텍스트의 성질을 파악하기 위해 사용

```python
fdist = FreqDist(text1) #FreqDist는 딕셔너리 타입
fdist #해당 단어가 몇번 출현했는지를 반환
```

```python
fdist.most_common(10) #가장 빈번한 것 10개를 반환
```

```python
fdist['that'] #'that이 해당 텍스트에서 몇번 출현했는지를 반환
#text1.count('that')과 동일한 기능
```

```python
fdist.freq('that') #that의 출현 확률을 반환
```

```python
fdist.plot(50) #상위 50개 단어의 출현 횟수를 그래프로 표현
```

```python
fdist.plot(50, cumulative=True) #상위 50개 단어의 출현 횟수를 이전 단어의 횟수에 누적해 그래프로 표현
```

```python
fdist.hapaxes() # 가장 빈번하지 않은(1번 출현한) 단어들을 반환
```

```python
#알파벳만 리스트에 추가
[word for word in fdist.hapaxes() if word.isalpha()]
```

```python
# 대소문자 구분 없이 freq dist 만들기
#set은 중복을 제거해 빈도수가 1이 되기 때문에 사용 x
fdist1 = FreqDist(text1) #the, The, THE를 각각
fdist2 = FreqDist([word.lower() for word in text1]) #the, The, THE를 한꺼번에
print(fdist1['the'])
print(fdist2['the'])
```

```python
#text2에서 길이가 3 이하이고 5번 이상 출현하는 단어만 리스트에 추가

#[word for word in text2 if len(word) <= 3 and text2.count(word) >= 5]
#text2의 단어를 모두 카운팅해야하기 때문에 비효율적임

[word for (word, count) in FreqDist(text2).items() if len(word) <= 3 and count >= 5]
#items()는 딕셔너리의 key와 value를 반환(각각 word, count로 들어감)
```

### for반복문
- 자료구조를 순회할 때 사용
- 특정 범위를 반복하면서 특정한 명령을 수행(범위는 range(), 문자열, 리스트 등으로 설정 가능)
- list comprehension: [<수행명령> for 변수 in <반복 범위>]
- range(start, end, step): 숫자를 순회할 때 사용
  - 인덱스와 동일한 규칙을 따르며 약간의 형태 차이가 존재(인덱스[start:end:step])

```python
#for반복문의 형태
#for <변수> in <반복 범위>:
#    <수행명령>
for word in sep_sent:
    print(word)
```

```python
#숫자는 range()를 이용
for i in range(1, 11):
    print(i)
```

```python
#zip()을 이용해 동시에 두 범위에서 두 변수를 가져올 수 있음
alphabet = 'abcdefg'
number = '1234567'

for a, n in zip(alphabet, number):
    print(a, n)
```

### Quiz 2
- text2에서 ch로 끝나는 단어를 중복 없이 출력
  - 첫 글자와 ch 남기고 가운데는 모두 -로 가리기
    - 예: much -> m_ch
    - ch로 시작하면 바꾸면 안됨
  - 정렬:
    - 문자 길이로 정렬

```python
words = list(set([word for word in text2 if word.endswith('ch')]))

cen_words = []

for word in words:
    if word.startswith('ch'):
        cen_words.append(word)
    else:
        cen_words.append(word[0] + (len(word)-3) * '_' + word[-2:])
        
cen_words.sort(key=len, reverse=True)

print(cen_words)
```
